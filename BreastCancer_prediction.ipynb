```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python


# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
import matplotlib.pyplot as plt

# Plotting training and validation accuracy
plt.figure(figsize=(10, 6))

# Plot training accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o', linestyle='-', color='b')

# Plot validation accuracy
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='s', linestyle='--', color='r')

# Adding title and labels
plt.title('Model Accuracy Over Epochs', fontsize=16)
plt.ylabel('Accuracy', fontsize=14)
plt.xlabel('Epoch', fontsize=14)

# Adding grid for better readability
plt.grid(True)

# Adding legend
plt.legend(['Training Data', 'Validation Data'], loc='lower right', fontsize=12)

# Adding a tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting training and validation loss
plt.figure(figsize=(10, 6))

# Plot training loss
plt.plot(history.history['loss'], label='Training Loss', marker='o', linestyle='-', color='b')

# Plot validation loss
plt.plot(history.history['val_loss'], label='Validation Loss', marker='s', linestyle='--', color='r')

# Adding title and labels
plt.title('Model Loss Over Epochs', fontsize=16)
plt.ylabel('Loss', fontsize=14)
plt.xlabel('Epoch', fontsize=14)

# Adding grid for better readability
plt.grid(True)

# Adding legend
plt.legend(['Training Data', 'Validation Data'], loc='upper right', fontsize=12)

# Adding a tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
 loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
import numpy as np

# Prompt the user for input data
input_string = input("Please enter the values for the following features, separated by spaces or commas:\n"
                     "radius_mean, texture_mean, perimeter_mean, area_mean, smoothness_mean, "
                     "compactness_mean, concavity_mean, concave_points_mean, symmetry_mean, fractal_dimension_mean, "
                     "radius_se, texture_se, perimeter_se, area_se, smoothness_se, "
                     "compactness_se, concavity_se, concave_points_se, symmetry_se, fractal_dimension_se, "
                     "radius_worst, texture_worst, perimeter_worst, area_worst, smoothness_worst, "
                     "compactness_worst, concavity_worst, concave_points_worst, symmetry_worst, fractal_dimension_worst\n")

# Split the input string into a list of strings and convert to floats
input_data = list(map(float, input_string.replace(',', ' ').split()))

# Convert input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

# Predicting the class using the trained model
prediction = model.predict(input_data_std)
print(prediction)

# Get the predicted class label (assuming the model output is a probability distribution)
prediction_label = np.argmax(prediction, axis=1)
print(prediction_label)

# Print the result
if prediction_label[0] == 0:
    print('The tumor is Malignant')
else:
    print('The tumor is Benign')
```
